```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Define the RBF function
def rbf(x, y, c, w):
    return np.exp(-((x - c[0]) ** 2 + (y - c[1]) ** 2) / (2 * w ** 2))

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Define the center and width of the RBF
center = [0, 0]  # Center of the RBF
width = 1.5      # Width of the RBF

# Calculate the RBF values for each point on the grid
Z = rbf(X, Y, center, width)

# Create a 2D heatmap with color mapping for the 4th dimension
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(Z, cmap=cm.viridis)

# Add a colorbar
cbar = fig.colorbar(cax)

# Set labels for the axes and a title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('4D Radial Basis Function Plot (Color Mapped)')

# Display the plot
plt.show()
```
